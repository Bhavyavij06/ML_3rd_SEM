# 100 random num in 1st quad

```python
import random 
qd1 = []
for i in range(1,100):
    qd1.append((random.uniform(2,4),random.uniform(2,4)))
```

# 100 random num in 3rd quad

```python
qd3 = []
for i in range(1,100):
    qd3.append((random.uniform(0,2),random.uniform(0,2)))
```

# Plot 

```python
# PLOT RANDOM NUMBERS 
import matplotlib.pyplot as plt
plt.scatter([i[0] for i in qd1] , [i[1] for i in qd1])
plt.scatter([i[0] for i in qd3] , [i[1] for i in qd3])
```

# Centroid

```python
# FINDING THE CENTER OF ALL THE POINTS IN QUADRANT 1 (USING MEAN) (X1+x2...x100/100,y1+y2...y100/100)
xqd1 = sum(i[0] for i in qd1)/len(qd1)
yqd1 = sum(i[1] for i in qd1)/len(qd1)
cqd1 = (xqd1 , yqd1)
cqd1
```

```python
# FINDING THE CENTER OF ALL THE POINTS IN QUADRANT 3 (USING MEAN) (X1+x2...x100/100,y1+y2...y100/100)
xqd3 = sum(i[0]  for i in qd3)/len(qd3)
yqd3 = sum(i[1]  for i in qd3)/len(qd3)
cqd3 = (xqd3 , yqd3)
cqd3
```

# Plot Centroids

```python
# PLOT THE CALCULATED CENTRES ON THE GRAPH
plt.scatter([i[0] for i in qd1] , [i[1] for i in qd1])
plt.scatter([i[0] for i in qd3] , [i[1] for i in qd3])
plt.scatter(xqd1 , yqd1 , color='m')
plt.scatter(xqd3 , yqd3 , color='c')
```

# Distance 

```python
dist_tst_cqd1 = ((xqd1 - x)**2 + (yqd1 - y)**2)**(1/2)
dist_tst_cqd3 = ((xqd3 - x)**2 + (yqd3 - y)**2)**(1/2)

distance = (dist_tst_cqd1 , dist_tst_cqd3)
distance
```

```python
if dist_tst_cqd1 >= dist_tst_cqd3:
    print("The Given Point Belongs To Cluster 2 (quadrant 3)")
else:
    print("The Given Point Belongs To Cluster 1 (quadrant 1)")
```
